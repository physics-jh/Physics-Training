### Plotting 기초(1)

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#prameter
s0 = 0
v0 = 2
g = 9.8

#equation
t = np.linspace(0,5,100)
s = s0 + v0*t + 0.5*g*t**2
v = v0 + g*t

#graph
fig, ax1 = plt.subplots()
ax1.plot(t, s)
fig, ax2 = plt.subplots()
ax2.plot(t, v)
plt.show()
```
