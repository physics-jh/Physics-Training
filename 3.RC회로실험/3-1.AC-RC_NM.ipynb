```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.integrate import odeint
```

```python
# ODE
f_l = 1
f_h = 50
t = np.linspace(0,3,300)
R = 100
C = 220e-6
V_in = 1.0 + 0.2*np.sin(2*np.pi*f_h*t) + 0.8*np.sin(2*np.pi*f_l*t)
def dq_dt(q, t):
    V_in = 1.0 + 0.2*np.sin(2*np.pi*f_h*t) + 0.8*np.sin(2*np.pi*f_l*t)

    return V_in/R - q/(R*C)
```

```python
Q = odeint(dq_dt, 0, t)
```

```python
fig, ax = plt.subplots()
ax.plot(t,V_in, label='V_in')
ax.plot(t,Q/C, label='V_filtered')
ax.set_xlabel('time')
ax.set_ylabel('V_C')
ax.legend()
ax.grid()
plt.show()
```
